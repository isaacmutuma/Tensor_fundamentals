```python
import torch
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
#using datsets you acess pre built data and
#transforms gives tools to process the data before feeding it to the model
# download FashionMNIST — the dataset you'll train on
training_data = datasets.FashionMNIST(
    root="data",#the downloaded files are saved in "data" folder
    train=True,# you just want the training portion on images
    download=True,# if files not on the computer download from the internet
    transform=transforms.ToTensor()#we convert the image into pytorch tensor
)
train_data_loader=DataLoader(
    training_data,
    batch_size=64,  #64, 1, 28, 28) batch size of 64 images with
                   #each image having a depth(greyscale is one feature)
                   #and each image has 28 by 28 pixels
    shuffle=True
)
#train_data give us our first batch, then iter makes sure they keep coming
#then next has the batches bundled up and they are then separated into image and labels
train_images,train_labels=next(iter(train_data_loader))
print("Batch image shape:", train_images.shape) #shape([64,1,28,28])
print("Batch image labels shape:",train_labels.shape) #shape([64])


#the progress bars are for showing the completed file downloads
print(len(training_data))        # should be 60000
print(training_data[0][0].shape) #the first item( a pair[0]) then access the image[0][0]
print(training_data[0][1])       # access the first item[0] then the items label[1]
```
