```python
import torch
import numpy as np
print(torch.__version__)

# creating tensors from data
data_one = [2, 3, 5, 6]
x = torch.tensor(data_one)  # 1D tensor and the shape is (4,) the number of elements

data_two = [[5, 6], [7, 8]]
y = torch.tensor(data_two)   # 2D tensor

# creating tensors from numpy
arr = np.array(data_one)
z = torch.from_numpy(arr)

# creating tensor from another tensor
w = torch.ones_like(x)

# creating tensors using shape
shape = (2, 3)
b = torch.zeros(shape)
c = torch.rand(shape)
d = torch.ones(shape)

# moving to GPU (only works if GPU available)
x = x.to('cuda' if torch.cuda.is_available() else 'cpu')

# numpy-like slicing and indexing
sl = torch.ones(4, 4)
rowone  = sl[0]
colone  = sl[:, 0]
lastcol = sl[:, -1]
lastrow = sl[-1]

# concatenating two tensors
f1 = torch.ones(2, 3)
f2 = torch.ones(2, 3)
f3 = torch.cat([f1, f2], dim=0)     # comma between f1 and f2

# matrix multiplication
f4 = f1 @ f2.T                      # dot product
f5 = f1 * f2                        # elementwise

# single element tensor
data = [3, 4, 5, 6, 7]
x = torch.tensor(data, dtype=torch.float32)
y = x.sum() # the number is still a pytorch object
print(y.item()) # the number is just a plain python number
#print("loss: " + str(loss.item()))   # works — loss.item() is a plain number

# in-place operation
x.add_(10)                           # underscore = in-place

# tensor to numpy
t = torch.tensor([1.0, 2.0, 3.0]) #initilaized our tensor
n = t.numpy()  # from t create a numpy array and call it n
t.add_(5) # notice the _ after that ensures theres shared memory between t and n otherwise the memory is unchanged
#t.add(5) # the only change here is on t not on n
print(n)

# numpy to tensor
arr2 = np.array([1.0, 2.0, 3.0])
v = torch.from_numpy(arr2)
np.add(arr2, 10, out=arr2)           # in-place numpy change
print(v)                             # v changes too



```
